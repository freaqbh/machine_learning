```python
import pandas as pd

# Load the uploaded dataset
file_path = '/content/drive/MyDrive/Colab Notebooks/data_ekonomi.csv'
data = pd.read_csv(file_path)

# Display basic info and first few rows
data.info(), data.head()
```

```python
# Strip whitespace from column names
data.columns = data.columns.str.strip()

# Cek jumlah missing values per kolom
missing_counts = data.isnull().sum()

# Tampilkan kolom dengan nilai null
missing_counts[missing_counts > 0]
```

# **Data Preprocessing**

```python
# Hapus kolom 'Perubahan_Inventaris' karena terlalu banyak missing values
data_cleaned = data.drop(columns=['Perubahan_Inventaris'])

# Imputasi missing values dengan median untuk kolom numerik lainnya
for col in data_cleaned.select_dtypes(include=['float64', 'int64']).columns:
    if data_cleaned[col].isnull().sum() > 0:
        median_value = data_cleaned[col].median()
        data_cleaned[col].fillna(median_value, inplace=True)

# Verifikasi bahwa tidak ada missing value tersisa
data_cleaned.isnull().sum().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

# Salin data untuk proses encoding
data_encoded = data_cleaned.copy()

# Identifikasi kolom kategorikal
categorical_cols = data_encoded.select_dtypes(include=['object']).columns

# Terapkan Label Encoding ke setiap kolom kategorikal
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    data_encoded[col] = le.fit_transform(data_encoded[col])
    label_encoders[col] = le

# Tampilkan hasil encoding untuk kolom kategorikal
data_encoded[categorical_cols].head()
```

# **Pemilihan Jumlah Cluster**

```python
features = [
    'PNB_Per_Capita',
    'Produk_Domestik_Bruto',
    'Penduduk',
    'Nilai_Tukar_IMF',
    'Ekspor_Barang_Jasa',
    'Impor_barang_jasa',
    'Pengeluaran_konsumsi_rumah_tangga',
    'Pengeluaran_Konsumsi_Akhir_Pemerintahan',
    'Industri_Pengolahan',
    'Pertanian, perburuan, kehutanan, perikanan'
]

X = data_encoded[features]

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

inertia = []
silhouette_scores = []
K = range(2, 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Plot Elbow
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(K, inertia, 'bo-')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Inertia')
plt.title('Metode Elbow')

# Plot Silhouette Score
plt.subplot(1, 2, 2)
plt.plot(K, silhouette_scores, 'ro-')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score per Cluster')
plt.tight_layout()
plt.show()
```

Menentukan Jumlah Cluster Optimal
Dalam praktik unsupervised learning (terutama clustering), dua teknik umum digunakan untuk mengevaluasi jumlah cluster terbaik:

a. Elbow Method
Mengamati nilai WCSS (Within-Cluster Sum of Squares) untuk berbagai jumlah cluster.

Kita mencari “tekukan” (elbow) di grafik—yakni titik di mana penambahan cluster berikutnya tidak banyak mengurangi WCSS.

Misalnya, jika grafik WCSS mulai melandai setelah k = 3, maka 3 kemungkinan jumlah cluster yang optimal.

b. Silhouette Score
Mengukur seberapa mirip sebuah titik dengan cluster-nya sendiri dibanding dengan cluster lain.

Skor mendekati 1.0 berarti cluster yang jelas terbentuk.

Jumlah cluster dengan silhouette score tertinggi biasanya dipilih.

Dalam eksperimen nyata, misalnya:

Elbow terjadi di k = 4

Silhouette Score tertinggi juga di k = 3 atau 4

k = 4 dipilih karena seimbang antara kompak dan terpisahnya cluster.

```python
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Tambahkan hasil cluster ke dataframe
data_encoded['Cluster'] = clusters
data_cleaned['Cluster'] = clusters

data_encoded.groupby('Cluster')[features].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

# Scatter plot berdasarkan dua fitur utama
sns.scatterplot(
    data=data_encoded,
    x='PNB_Per_Capita',
    y='Produk_Domestik_Bruto',
    hue='Cluster',
    palette='Set2',
    s=60,
    alpha=0.7
)

plt.title('Visualisasi Clustering Negara Berdasarkan Indikator Ekonomi')
plt.xlabel('PNB Per Kapita')
plt.ylabel('Produk Domestik Bruto')
plt.grid(True)
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()

# Gabungkan informasi negara dan cluster
clustered_countries = data_cleaned[['Negara', 'Tahun', 'Cluster']]

# Tampilkan beberapa negara dari setiap cluster
for cluster_id in sorted(clustered_countries['Cluster'].unique()):
    print(f"\nCluster {cluster_id}:")
    negara_cluster = clustered_countries[clustered_countries['Cluster'] == cluster_id]['Negara'].unique()
    print(", ".join(negara_cluster[:15]))  # tampilkan max 15 negara dulu
    print(f"... total: {len(negara_cluster)} negara")
```

# **Evaluasi Internal**

```python
sil_score = silhouette_score(X_scaled, data_encoded['Cluster'])
print(f"Silhouette Score: {sil_score:.4f}")

from sklearn.metrics import davies_bouldin_score

db_score = davies_bouldin_score(X_scaled, data_encoded['Cluster'])
print(f"Davies-Bouldin Index: {db_score:.4f}")

from sklearn.metrics import calinski_harabasz_score

ch_score = calinski_harabasz_score(X_scaled, data_encoded['Cluster'])
print(f"Calinski-Harabasz Score: {ch_score:.2f}")
```
